## Menentukan Pertanyaan Bisnis

- Bagaimana pengaruh musim terhadap jumlah penyewaan sepeda?
- Bagaimana pola penggunaan sepeda berdasarkan waktu?

## Import Semua Packages/Library yang Digunakan

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
bikes_df = pd.read_csv("./data/bike_hour.csv")
bikes_df.head()
```

**Insight:**
- Terdapat 17 column

### Assessing Data

```python
bikes_df.info()

print("Jumlah duplikasi: ", bikes_df.duplicated().sum())
```

**Insight:**
- jumlah data pada bikes_df telah lengkap sehingga tidak ada missing value di dalamnya
- kolom dteday harusnya direpresentasikan sebagai tipe data **datetime**, bukan **object**
- tidak ada duplikasi data
- tidak ada keanehan pada data

### Cleaning Data

```python
datetime_columns = ["dteday"]

for column in datetime_columns:
  bikes_df[column] = pd.to_datetime(bikes_df[column])
```

```python
bikes_df.info()
```

**Insight:**
- mengganti tipe data pada kolom dteday menjadi datetime

## Exploratory Data Analysis (EDA)

### Explore ...

```python
print("Total casual: ", bikes_df.casual.sum())
print("Total registered: ", bikes_df.registered.sum())

bikes_df.describe(include="all")
```

**Insight:**
- jumlah pengguna yang terdaftar lebih banyak dari casual dengan rata rata pengguna yang terdaftar(registered) adalah **153,79**, dimana pengguna casual hanya sebesar **35.67** per jamnya
- rentang bulan dari 1 sampai 12 yang menunjukkan bahwa dataset ini mencakup seluruh bulan
- rentang hour dari 0 sampai 23 yang menunjukkan bahwa dataset mencakup seluruh jam

```python
bikes_df["season"] = bikes_df.season.apply(lambda x: "spring" if x == 1 else ("summer" if x == 2 else ("fall" if x == 3 else "winter")))
bikes_df.groupby(by=["yr","season"]).agg({
    "casual": "sum",
    "registered": "sum",
    "cnt": "sum"
})
```

**Insight:**
- dapat terlihat **musim fall** merupakan penyewaan tertinggi dan **season ke 1** merupakan penyewaan terendah

```python
bikes_df["hr_group"] = bikes_df.hr.apply(lambda x: "Morning" if  3 <= x < 11 else ("Midday" if  11 <= x < 18 else "Night"))
bikes_df.groupby(by="hr_group").cnt.sum().sort_values(ascending=False).reset_index()
```

```python
bikes_df.groupby(by=["hr_group","hr"]).agg({
    "casual": "sum",
    "registered": "sum",
    "cnt": "sum"
}).sort_values(by="hr")
```

**Insight:**
- dapat terlihat penyewaan tertinggi berada pada **Midday** yaitu pukul 11:00 - 17:00 dan penyewaan terendah berada pada **Morning** yaitu pukul 3:00 - 10:00
- penyewaan tertinggi berada pada pukul **17:00** dan terendah pukul **4:00**

## Visualization & Explanatory Analysis

### Pertanyaan 1: 

```python
byseason_df = bikes_df.groupby(by="season").cnt.sum().reset_index()

byseason_df.rename(columns={
    "cnt": "rent_count"
}, inplace=True)

plt.figure(figsize=(10, 5))
colors_ = ["#D3D3D3", "#D3D3D3", "#D3D3D3", "#72BCD4"]

sns.barplot(
    y="rent_count", 
    x="season",
    hue="season",
    data=byseason_df.sort_values(by="season", ascending=False),
    palette=colors_,
    legend=False
)

plt.title("Number of Rent by Season", loc="center", fontsize=15)
plt.ylabel("total_rent")
plt.xlabel("season")
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

```python
hourly_rent_df = bikes_df.groupby(by="hr").cnt.sum().reset_index()

hourly_rent_df.rename(columns={
    "hr": "hour",
    "cnt": "count_rent"
}, inplace=True)

plt.figure(figsize=(10, 5)) 
plt.xlabel("Hour")
plt.ylabel("total_rent")
plt.plot(hourly_rent_df["hour"], hourly_rent_df["count_rent"], marker='o', linewidth=2, color="#72BCD4") 
plt.title("Number of Rent per Hour(24h)", loc="center", fontsize=15) 
plt.xticks(hourly_rent_df["hour"], fontsize=10) 
plt.yticks(fontsize=10) 
plt.show()
```

**Insight:**
- berdasarkan season dapat dilihat bahwa **season ke 3** merupakan penyewaan tertinggi
- berdasarkan jumlah sewa per jam dapat dilihat bahwa **jam 17** merupakah penyewaan tertinggi

## Conclusion

- Musim memiliki dampak yang signifikan terhadap jumlah penyewaan sepeda. Berdasarkan analisis, **musim 3 (fall)** merupakan musim dengan rata-rata penyewaan sepeda tertinggi dibandingkan musim lainnya. Hal ini kemungkinan disebabkan oleh kondisi cuaca yang lebih ideal untuk bersepeda pada musim ini, dengan suhu yang cenderung nyaman dan cuaca yang lebih stabil.
  
- Analisis berdasarkan waktu juga menunjukkan bahwa jam **17.00 (5 sore)** adalah puncak penyewaan sepeda harian. Ini menunjukkan pola penggunaan yang kuat di sekitar jam pulang kerja, di mana banyak pengguna yang menggunakan sepeda untuk pulang dari tempat kerja atau aktivitas sehari-hari.
